```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
import sqlite3

conn = sqlite3.connect("/Users/airkumar/jaw.db")
```

```python
df = pd.read_sql_query("select * from datatable;", conn)
```

```python
df['reading'] = pd.to_numeric(df['reading'])
```

```python
df['recorded'] = pd.to_datetime(df['recorded'])
```

```python
x = df['recorded']
y = df['reading']
start_date = '2020-02-23 06:00:00'
end_date = '2020-02-23 15:00:00'
xTodayFilter = x > start_date
filteredDf = df.loc[xTodayFilter]
xFilter = filteredDf['recorded']
yFilter = filteredDf['reading']
filteredDf.head()
```

```python
plt.scatter(xFilter,yFilter,s=1)
plt.xlabel("Time")
plt.ylabel("Sensor Reading")
plt.xlim(start_date,end_date)

plt.show()
```
